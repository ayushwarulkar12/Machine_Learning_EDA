```python
# =============================
#  NYC Motor Vehicle Collisions EDA
# =============================

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# Step 2: Load Dataset
df = pd.read_csv("Motor_Vehicle_Collisions_-_Crashes.csv")
print("Shape of dataset:", df.shape)
print("\nColumns:\n", df.columns)

# Step 3: Basic Info
print("\nData Types:\n")
print(df.dtypes)
print("\nMissing Values:\n")
print(df.isnull().sum().sort_values(ascending=False).head(10))

# Step 4: Preview Data
print("\nSample Data:")
print(df.head())

# Step 5: Summary Statistics
print("\nNumerical Summary:")
print(df.describe())

print("\nCategorical Summary:")
print(df.describe(include='object'))

# Step 6: Data Cleaning
# Convert date/time columns
if 'CRASH DATE' in df.columns:
    df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'], errors='coerce')
if 'CRASH TIME' in df.columns:
    df['CRASH TIME'] = pd.to_datetime(df['CRASH TIME'], errors='coerce', format='%H:%M')

# Extract useful time features
df['YEAR'] = df['CRASH DATE'].dt.year
df['MONTH'] = df['CRASH DATE'].dt.month
df['DAY_OF_WEEK'] = df['CRASH DATE'].dt.day_name()
df['HOUR'] = df['CRASH TIME'].dt.hour

# Step 7: Crashes Over Time
plt.figure(figsize=(10,5))
df['YEAR'].value_counts().sort_index().plot(kind='bar', color='steelblue')
plt.title('Number of Crashes per Year')
plt.xlabel('Year')
plt.ylabel('Crashes')
plt.show()

# Step 8: Crashes by Month
plt.figure(figsize=(10,5))
df['MONTH'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Crashes by Month')
plt.xlabel('Month')
plt.ylabel('Number of Crashes')
plt.show()

# Step 9: Crashes by Day of Week
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='DAY_OF_WEEK', order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title('Crashes by Day of the Week')
plt.xticks(rotation=45)
plt.show()

# Step 10: Crashes by Hour
plt.figure(figsize=(10,5))
sns.histplot(df['HOUR'].dropna(), bins=24, kde=False, color='green')
plt.title('Crashes by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Crashes Count')
plt.show()

# Step 11: Borough Analysis
if 'BOROUGH' in df.columns:
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, y='BOROUGH', order=df['BOROUGH'].value_counts().index, palette='viridis')
    plt.title('Crashes by Borough')
    plt.ylabel('Borough')
    plt.xlabel('Number of Crashes')
    plt.show()

# Step 12: Injury Analysis
injury_cols = ['NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
               'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF CYCLIST INJURED',
               'NUMBER OF MOTORIST INJURED']

for col in injury_cols:
    if col in df.columns:
        plt.figure(figsize=(6,4))
        sns.histplot(df[col], bins=20, kde=False, color='crimson')
        plt.title(f'Distribution of {col}')
        plt.show()

# Step 13: Top Contributing Factors
factor_cols = [col for col in df.columns if 'CONTRIBUTING FACTOR' in col]
for col in factor_cols:
    top_factors = df[col].value_counts().head(10)
    plt.figure(figsize=(8,4))
    sns.barplot(x=top_factors.values, y=top_factors.index, palette='coolwarm')
    plt.title(f'Top 10 {col}')
    plt.xlabel('Count')
    plt.ylabel('Factor')
    plt.show()

# Step 14: Vehicle Type Analysis
vehicle_cols = [col for col in df.columns if 'VEHICLE TYPE' in col]
for col in vehicle_cols:
    top_types = df[col].value_counts().head(10)
    plt.figure(figsize=(8,4))
    sns.barplot(x=top_types.values, y=top_types.index, palette='magma')
    plt.title(f'Top 10 {col}')
    plt.xlabel('Count')
    plt.ylabel('Vehicle Type')
    plt.show()

# Step 15: Correlation Heatmap
num_cols = df.select_dtypes(include=np.number)
plt.figure(figsize=(10,6))
sns.heatmap(num_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.upper()

# Encode categorical feature 'BOROUGH'
le = LabelEncoder()
df['BOROUGH'] = le.fit_transform(df['BOROUGH'].astype(str))

# Define features and target
X = df[['BOROUGH', 'NUMBER OF PERSONS KILLED',
        'NUMBER OF PEDESTRIANS INJURED',
        'NUMBER OF CYCLIST INJURED',
        'NUMBER OF MOTORIST INJURED']]

y = df['INJURY_FLAG']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 8: Train and Evaluate Models

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
print("Logistic Regression Performance:")
print("Accuracy:", accuracy_score(y_test, lr_pred))
print(classification_report(y_test, lr_pred))

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
print("\nRandom Forest Performance:")
print("Accuracy:", accuracy_score(y_test, rf_pred))
print(classification_report(y_test, rf_pred))

# Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb.fit(X_train, y_train)
gb_pred = gb.predict(X_test)
print("\nGradient Boosting Performance:")
print("Accuracy:", accuracy_score(y_test, gb_pred))
print(classification_report(y_test, gb_pred))

# Neural Network (using a simple MLP)
from sklearn.neural_network import MLPClassifier

# Scale features for Neural Network
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


mlp = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
mlp.fit(X_train_scaled, y_train)
mlp_pred = mlp.predict(X_test_scaled)
print("\nNeural Network Performance:")
print("Accuracy:", accuracy_score(y_test, mlp_pred))
print(classification_report(y_test, mlp_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Read your uploaded CSV
df = pd.read_csv("/content/Motor_Vehicle_Collisions_-_Crashes.csv")

# Keep only numeric columns and drop rows with NaN
numeric_df = df.select_dtypes(include=[np.number]).dropna()

# Optional: sample smaller subset to prevent memory crash
numeric_df = numeric_df.sample(n=20000, random_state=42)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(numeric_df)

print(" Data ready for clustering:", X_scaled.shape)
```

K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Find optimal K using Elbow Method
inertia = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.figure(figsize=(8,4))
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for K-Means")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()

# Choose K=4 (you can adjust after elbow plot)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)
numeric_df["KMeans_Cluster"] = kmeans_labels

print("\n K-Means Cluster Distribution:")
print(numeric_df["KMeans_Cluster"].value_counts())

# 2D Visualization using PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=kmeans_labels, palette="viridis")
plt.title("K-Means Clusters (PCA)")
plt.show()
```

DBSCAN CLUSTERING

```python
from sklearn.cluster import DBSCAN

# Adjust eps and min_samples if needed
dbscan = DBSCAN(eps=1.5, min_samples=10)
dbscan_labels = dbscan.fit_predict(X_scaled)
numeric_df["DBSCAN_Cluster"] = dbscan_labels

print("\n DBSCAN Cluster Distribution:")
print(numeric_df["DBSCAN_Cluster"].value_counts())

# PCA Visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=dbscan_labels, palette="Set2")
plt.title("DBSCAN Clusters (PCA)")
plt.show()
```

HIERARCHICAL CLUSTERING

```python
from sklearn.cluster import AgglomerativeClustering

# Try fewer samples if its slow
sample_for_hier = numeric_df.sample(n=5000, random_state=42)
X_hier = scaler.fit_transform(sample_for_hier)

hier = AgglomerativeClustering(n_clusters=4)
hier_labels = hier.fit_predict(X_hier)
sample_for_hier["Hierarchical_Cluster"] = hier_labels

print("\n Hierarchical Cluster Distribution:")
print(sample_for_hier["Hierarchical_Cluster"].value_counts())

# PCA Visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_hier)

sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=hier_labels, palette="tab10")
plt.title("Hierarchical Clusters (PCA)")
plt.show()
```

Prophet Forecasting

```python
from prophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt

# Prepare data (if not done already)
df = pd.read_csv("/content/Motor_Vehicle_Collisions_-_Crashes.csv")
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'], errors='coerce')
daily_crashes = df.groupby('CRASH DATE').size().reset_index(name='crash_count')
daily_crashes = daily_crashes.rename(columns={'CRASH DATE': 'ds', 'crash_count': 'y'})

# Fit Prophet model
model = Prophet(daily_seasonality=True, yearly_seasonality=True)
model.fit(daily_crashes)

# Forecast next 90 days
future = model.make_future_dataframe(periods=90)
forecast = model.predict(future)

# Plot
model.plot(forecast)
plt.title(" Daily Crash Forecast (Prophet)")
plt.show()

# Components (trend, seasonality)
model.plot_components(forecast)
plt.show()
```

```python
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Load and clean
df = pd.read_csv("/content/Motor_Vehicle_Collisions_-_Crashes.csv")
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'], errors='coerce')
df = df.dropna(subset=['BOROUGH', 'CRASH DATE'])

# Sort and compute time gaps per borough
df = df.sort_values(['BOROUGH', 'CRASH DATE'])
df['time_since_prev'] = df.groupby('BOROUGH')['CRASH DATE'].diff().dt.days

# Remove NaNs (first crash in each borough)
df = df.dropna(subset=['time_since_prev'])
df['event'] = 1  # every interval ends with a crash

# Fit Kaplan-Meier model
km = KaplanMeierFitter()

plt.figure(figsize=(8,6))
for borough, group in df.groupby('BOROUGH'):
    km.fit(durations=group['time_since_prev'], event_observed=group['event'], label=borough)
    km.plot_survival_function()

plt.title("KaplanMeier Survival Curves  Time Between Crashes per Borough")
plt.xlabel("Days Since Previous Crash")
plt.ylabel("Probability of No Crash")
plt.legend()
plt.show()
```
